```python
#Import library
import pandas as pd
import time
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn import preprocessing
```

```python
#1. Import our training and testing dataset
df_training = pd.read_csv("data/train.csv",header=0,index_col='id')
print("size of training data:"+str(len(df_training)))
df_testing = pd.read_csv("data/test.csv",header=0,index_col=None)
print("size of testing data:"+str(len(df_testing)))
```

```python
df_training
```

```python
df_testing
```

```python
# using dropna() function to drop rows with at least one column with null value
df_training.dropna(inplace=True)
```

```python
X=df_training[['MO HLADR+ MFI (cells/ul)','Neu CD64+MFI (cells/ul)','CD3+T (cells/ul)','CD8+T (cells/ul)','CD4+T (cells/ul)','NK (cells/ul)','CD19+ (cells/ul)','CD45+ (cells/ul)','NK (cells/ul)','Age','Sex 0M1F','Mono CD64+MFI (cells/ul)']]  # Features
y=df_training['label']  # Labels
```

```python
X
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
X_test
```

```python
#Create a Random Forest Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Predict test data
test_X=df_testing[['MO HLADR+ MFI (cells/ul)','Neu CD64+MFI (cells/ul)','CD3+T (cells/ul)','CD8+T (cells/ul)','CD4+T (cells/ul)','NK (cells/ul)','CD19+ (cells/ul)','CD45+ (cells/ul)','NK (cells/ul)','Age','Sex 0M1F','Mono CD64+MFI (cells/ul)']]  # Features
test_pred=clf.predict(test_X)
```

```python
test_pred
```

```python
#Export result to submission csv file
submission = pd.DataFrame({'id': df_testing.id, 'label': [int(i) for i in test_pred]})
submission.head()
submission.to_csv('submission.csv', index=False)
```
